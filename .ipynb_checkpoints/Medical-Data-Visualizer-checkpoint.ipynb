```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
patients = pd.read_csv("medical_examination.csv")
```

```python
patients.head()
```

```python
patients.columns
```

### 1. Add an `overweight` column to the data. 

#### To determine if a person is overweight, first calculate their BMI by dividing their weight in kilograms by the square of their height in meters. 

#### If that value is > 25 then the person is overweight. Use the value 0 for NOT overweight and the value 1 for overweight.
![purple-divider](https://user-images.githubusercontent.com/7065401/52071927-c1cd7100-2562-11e9-908a-dde91ba14e59.png)

```python
toMeters = patients['height'] / 100
bmi = patients['weight'] / (toMeters ** 2)

# new column based on if statement using np.where()
patients['overweight'] = np.where(bmi>25, 1, 0)

patients
```

### 2. Normalize the data by making 0 always good and 1 always bad. 

#### If the value of `cholesterol` or `gluc` is 1, make the value 0. If the value is more than 1, make the value 1.
![purple-divider](https://user-images.githubusercontent.com/7065401/52071927-c1cd7100-2562-11e9-908a-dde91ba14e59.png)

```python
patients['cholesterol'] = np.where(patients['cholesterol']==1, 0, 1)
patients['gluc'] = np.where(patients['gluc']==1, 0, 1)

patients
```

### 3. Convert the data into long format and create a chart that shows the value counts of the categorical features using seaborn's `catplot()`. 

#### The dataset should be split by 'Cardio' so there is one chart for each `cardio` value. 

#### The chart should look like `examples/Figure_1.png`.
![purple-divider](https://user-images.githubusercontent.com/7065401/52071927-c1cd7100-2562-11e9-908a-dde91ba14e59.png)

```python
columns = [
      'active',
      'alco',
      'cholesterol',
      'gluc',
      'overweight',
      'smoke'
]
    
# melt() converts dataframe from wide format to long format
patients_cat = pd.melt(patients, id_vars=["cardio"], value_vars=columns)
    
# Group and reformat the data to split it by 'cardio'. Show the counts of each feature.
patients_cat = patients_cat.reset_index().groupby(['variable', 'cardio', 'value']).agg('count').rename(columns={'index': 'total'}).reset_index()

# Draw the catplot with 'sns.catplot()'
# fig.fig (?)
fig = sns.catplot(
    x="variable",
    y="total",
    col="cardio",
    hue="value",
    data=patients_cat,
    kind="bar").fig
    
    
plt.savefig('catplot.png')
```

### 4. Clean the data. Filter out the following patient segments that represent incorrect data (EXCLUDE the following):
![purple-divider](https://user-images.githubusercontent.com/7065401/52071927-c1cd7100-2562-11e9-908a-dde91ba14e59.png)

##### *a. diastolic pressure is higher than systolic (Keep the correct data with `(patients['ap_lo'] <= patients['ap_hi'])`)*

```python
(patients['ap_lo'] <= patients['ap_hi']) # include in filter; opposite of above statement
```

##### *b. height is less than the 2.5th percentile  (Keep the correct data with `(patients['height'] >= patients['height'].quantile(0.025))`)*

```python
(patients['height'] >= patients['height'].quantile(0.025)) # include in filter; opposite of above statement
```

##### *c. height is more than the 97.5th percentile*

```python
patients['height'] <= patients['height'].quantile(0.975) # include in filter; opposite of above statement
```

##### *d. weight is less than the 2.5th percentile*

```python
patients['weight'] >= patients['weight'].quantile(0.025) # include in filter; opposite of above statement
```

##### *e. weight is more than the 97.5th percentile*

```python
patients['height'] <= patients['height'].quantile(0.975) # include in filter; opposite of above statement
```

#### FINAL RESULT

```python
patients_heatmap = patients[
      (patients['ap_lo'] <= patients['ap_hi'])
      & (patients['height'] >= patients['height'].quantile(0.025))
      & (patients['height'] <= patients['height'].quantile(0.975))
      & (patients['weight'] >= patients['weight'].quantile(0.025))
      & (patients['weight'] <= patients['weight'].quantile(0.975))
    ]
```

### 5. Create a correlation matrix using the dataset. Plot the correlation matrix using seaborn's `heatmap()`. 

#### Mask the upper triangle. 

#### The chart should look like `examples/Figure_2.png`.
![purple-divider](https://user-images.githubusercontent.com/7065401/52071927-c1cd7100-2562-11e9-908a-dde91ba14e59.png)

#### np.triu(): Select values in array to make a diagonal triangle



```python
correlations = patients_heatmap.corr()

#Mask for upper part of the triangle (?)
mask = np.triu(correlations)
mask[np.triu_indices_from(mask)] = True

fig, ax = plt.subplots(figsize=(12, 9))

#blue-black-red color palette
#colormap = sns.diverging_palette(250, 359, s=75, l=40, n=40, center="dark")
#cmap='icefire'

#triangle heatmap graph
ax = sns.heatmap(
        correlations,
        mask=mask,
        vmax=0.3,
        vmin=-0.10,
        center=0,
        square=True,
        fmt=".1f",
        annot=True,
    )

fig.savefig('heatmap.png')
```
